```python
#Import the libraries
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data3= pd.read_csv("C:/Users/abc/Desktop/Assignments/Assignment 7/crime_data.csv")
data3.head()
```

```python
data3.shape
```

```python
data3.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
array=data3.values
```

```python
stscaler = StandardScaler().fit(array)
X = stscaler.transform(array)
```

```python
#min sample is equal to totao feacture +1
dbscan = DBSCAN(eps=1.3, min_samples=5)
dbscan.fit(X)
```

```python
#Noisy samples are given the label -1.
dbscan.labels_
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['cluster'])
```

```python
pd.concat([data3,cl],axis=1)
```
